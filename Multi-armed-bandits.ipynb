### Multi Armed Bandit Problem Definition 

**Core Challenge**

- The fundamental challenge lies in balancing exploration and exploitation

**Trade-off**

- Exclusively choosing the best known option risks missing the optimal solution
- Constantly trying new options leads to frequent suboptimal experiences

### Bernoulli Multi Armed Bandit 

**Core Concept**

- Multi-armed bandits deal with the dilemma between exploitation of known rewards and exploration of new options under incomplete information
- The goal is to balance utilizing proven options while taking calculated risks to maximize long-term rewards

**Bernoulli Multi Armed Bandits**

- System consists of K arms with different reward distributions (Q1, Q2, QK)
- The process involves a set of actions and their corresponding rewards (A, R)
- Expected rewards are based on the probability of success in Bernoulli trials

**Action Selection Methods**

- Three main selection strategies exist:
    - Random/greedy selection
    - Epsilon greedy selection
    - Upper confidence and Thompson selection
